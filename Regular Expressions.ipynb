```python
#11) 

import re

def matchstring(text):
  
    pattern = r'^[A-Za-z0-9_]+$'
    if re.fullmatch(pattern, text):
        return True
    else:
        return False

teststrings = [
    "Valid-String123",
    "Invalid-String!",
    "Another-valid-string_456",
    "Invalid string with spaces",
    "Valid-123"
]

for string in teststrings:
    result = matchstring(string)
    print(f"'{string}' is valid: {result}")

```

```python
#12) 

import re

def startswithnumber(text, number):
    pattern = re.compile(rf'^{number}')
    
    if pattern.match(text):
        return True
    else:
        return False

test_strings = [
    "123abc",
    "456def",
    "789ghi",
    "abc123",
    "123def456"
]

specificnumber = "123"

for string in test_strings:
    result = startswithnumber(string, specificnumber)
    print(f"'{string}' starts with '{specificnumber}': {result}")


```

```python
#13)

import re

def removeleadingzeros(ipaddress):
    pattern = re.compile(r'\b0+(\d+)\b')
    cleanedip = pattern.sub(r'\1', ipaddress)
    return cleanedip

ipaddresses = [
    "192.168.001.001",
    "010.020.030.040",
    "001.002.003.004",
    "123.045.067.089"
]

cleanedipaddresses = [removeleadingzeros(ip) for ip in ipaddresses]
for original, cleaned in zip(ipaddresses, cleanedipaddresses):
    print(f"Original: {original} -> Cleaned: {cleaned}")

```

```python
 #19) 
def convertdateformat(datestr):
    parts = datestr.split('-')
    
    if len(parts) == 3:
        newdatestr = f"{parts[2]}-{parts[1]}-{parts[0]}"
        return newdatestr
    else:
        return "Invalid date format"

datestr = "2023-05-25"
converteddate = convertdateformat(datestr)
print(f"Original date: {datestr}")
print(f"Converted date: {converteddate}")

```

```python
#22)

import re

def findmaxnumericvalue(text):
    pattern = re.compile(r'\d+')
    matches = pattern.findall(text)
    numericvalues = [int(match) for match in matches]

    if numericvalues:
        return max(numericvalues)
    else:
        return None
sampletext = 'My marks in each semester are: 947, 896, 926, 524, 734, 950, 642'
maxvalue = findmaxnumericvalue(sampletext)

print(f"The maximum numeric value is: {maxvalue}")

```
